```python
import matplotlib.pyplot as plt
import numpy as np

types = ['Unprocessed', 'Undersampled', 'Augmented', 'Sub Images']
Anomaly = [100, 100, 500, 375]
Normal = [500, 100,500, 375]

bar_width = 0.35

r1 = np.arange(len(types))
r2 = [x + bar_width for x in r1]

plt.bar(r1, Anomaly, color='#5A67C8', width=bar_width, edgecolor='grey', label='Anomaly')
plt.bar(r2, Normal, color='#E68545', width=bar_width, edgecolor='grey', label='Normal')

plt.xlabel('Technique', fontweight='bold')
plt.ylabel('Number of Images', fontweight='bold')
plt.title('Data pre-processing')
plt.xticks([r + bar_width / 2 for r in range(len(types))], types)
plt.legend()

# Show the plot
plt.show()
```
