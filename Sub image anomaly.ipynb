```python
import cv2
import os

from google.colab import drive

# drive.mount('/content/drive')

src = "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/Augmented/Anomaly"
dst = "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/CroppedAnomaly"

def dst_im_path(cnt):
  if cnt <= 900:
    return "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/CroppedAnomaly/surah"
  elif cnt <= 1800:
    return "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/CroppedAnomaly/shikha"
  elif cnt <= 2700:
    return "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/CroppedAnomaly/htoo"
  elif cnt <= 3600:
    return "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/CroppedAnomaly/rabbi"
  else:
    return "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/CroppedAnomaly/asif"

file_counter = 0

for filename in os.listdir(src):
  image_path = os.path.join(src, filename)
  print('processing', image_path)
  image = cv2.imread(image_path)

  height, width = image.shape[:2]
  part_height = height // 3
  part_width = width // 3

  for i in range(3):
    for j in range(3):
      part = image[i * part_height: (i + 1) * part_height, j * part_width: (j + 1) * part_width]
      file_counter = file_counter + 1
      dst = dst_im_path(file_counter)
      fname = str(file_counter) + ".JPG"
      output_path = os.path.join(dst, fname)
      print('saving',output_path)
      cv2.imwrite(output_path, part)
```
