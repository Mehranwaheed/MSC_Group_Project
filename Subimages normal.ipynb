```python
import cv2
import os

from google.colab import drive

drive.mount('/content/drive')

src = "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/Augmented/Normal"
dst = "/content/drive/MyDrive/Colab Notebooks/faulty-product-detection/Data/All Subimages/Normal"

file_counter = 0

for filename in os.listdir(src):
  image_path = os.path.join(src, filename)
  # print('processing', image_path)
  image = cv2.imread(image_path)

  height, width = image.shape[:2]
  part_height = height // 3
  part_width = width // 3

  for i in range(3):
    for j in range(3):
      part = image[i * part_height: (i + 1) * part_height, j * part_width: (j + 1) * part_width]
      file_counter = file_counter + 1

      fname = str(file_counter) + ".JPG"
      output_path = os.path.join(dst, fname)
      # print('saving',output_path)
      cv2.imwrite(output_path, part)
  if file_counter >= 500 :
    break

```
